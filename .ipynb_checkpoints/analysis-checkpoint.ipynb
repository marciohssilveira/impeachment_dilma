# Brazil's 2016 impeachment process analysis
The year is 2016, the majority of the federal deputies in Brazil voted 'yes' for the impeachment of Dilma Roussef, president at the time. Some information of all deputies were provided by the government website, including the lines of the speech they gave during the votation process. <br><br>
In this notebook, I'll be analysing some voters characteristics such as gender, party and state of origin. Also, I'll be using some natural language processing for a more complex analysis of their speech. <br><br>
You'll see that the data is in brazilian portuguese language, some translation will be provided.

```python
#importing libraries for dealing with the datasets and presenting information in graphs

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#putting the csv data into a pandas dataframe

impeachment = pd.read_csv('impeachment_br.csv',index_col=0)
```

```python
impeachment.head()
```

```python
for col in impeachment.columns:
    if col not in ['Deputado', 'Fala']:
        print(f"{col}: {impeachment[col].unique()}")
```

```python
impeachment.Voto = impeachment.Voto.str.title()
```

```python
impeachment['Voto'].value_counts()
```

```python
def trata_voto(x):
    if x == 'Sim':
        return 1
    else:
        return 0
```

```python
imp = impeachment.copy()
```

```python
imp['Voto'] = imp['Voto'].apply(trata_voto)
```

```python
imp.head()
```

```python
((imp.groupby('Partido')['Voto'].sum() / imp.Partido.value_counts()).sort_values()*100).plot(kind='barh',
                                                                                             figsize=(5,10),
                                                                                             title='Percentage of YES parties')
```

```python
((imp.groupby('Estado')['Voto'].sum() / imp.Estado.value_counts()).sort_values()*100).plot(kind='barh',
                                                                                           figsize=(5,10),
                                                                                           title='Percentage of YES states')
```

Predicting the vote by party...

```python
y = imp.Voto
X = pd.get_dummies(imp[['Partido']])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
y_pred_S = pd.Series(data=y_pred, index = y_test.index, name='Predicted')
```

```python
results = pd.concat([y_pred_S, y_test, imp.Partido[y_test.index]],axis=1)
results.head()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
```

<a href='https://en.wikipedia.org/wiki/Confusion_matrix'>Confusion Matrix</a>

```python
matrix = confusion_matrix(y_test,y_pred)
matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
metrics = pd.DataFrame(columns=['k','Accuracy','Recall','Precision'])
```

```python
for k in range(1,30):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    matrix = confusion_matrix(y_test,y_pred)
    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred)
    recall = recall_score(y_test,y_pred)
    metrics.loc[k] = [k,accuracy,recall,precision]
```

```python
metrics.sort_values(by=['Recall','Precision','Accuracy'],ascending=False)
```

```python
ax = metrics.plot(kind='line',x='k',y='Recall')
metrics.plot(kind='line',x='k',y='Accuracy',ax=ax)
metrics.plot(kind='line',x='k',y='Precision',ax=ax)
```

# NTLK

```python
impeachment.head()
```

```python
sns.countplot(impeachment['Voto'])
```

The proportions of yes and no would imply in bias, let's sample by the size of the no, then...

```python
yes = impeachment.loc[impeachment['Voto'] == 'Sim']
no = impeachment.loc[impeachment['Voto'] == 'Não']
sample_yes = yes.sample(len(no))
```

Now let's create a sample dataframe with the same proportions of yes and no to avoid bias...

```python
sample_votes = pd.concat([no, sample_yes], axis=0)
```

```python
sample_votes.iloc[:2].Fala.values
```

```python
def add_space(x):
    return ' ' + x

sample_votes['Fala'] = sample_votes['Fala'].apply(add_space)
```

```python
sample_votes.iloc[:2].Fala.values
```

```python
yes = sample_votes.loc[sample_votes['Voto'] == 'Sim']
no = sample_votes.loc[sample_votes['Voto'] == 'Não']
```

```python
text_yes = yes['Fala'].sum()
text_no = no['Fala'].sum()
```

```python
import nltk as nl
#nl.download()
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import RSLPStemmer
from nltk.corpus import stopwords
nl.download('stopwords')
```

```python
token_yes = word_tokenize(text_yes)
token_no = word_tokenize(text_no)
```

```python
#retirando as stopwords e nao esquecendo de nao considerar pontos e simbolos len(palavra) > 1

list_yes = []
for word in token_yes:
    if word not in stopwords.words('Portuguese') and len(word) > 1:
        list_yes.append(word)

list_no = []
for word in token_no:
    if word not in stopwords.words('Portuguese') and len(word) > 1:
        list_no.append(word)
```

```python
series_yes = pd.Series(list_yes)
series_no = pd.Series(list_no)
```

```python
m_yes = series_yes.value_counts().head(30).sort_values()
m_no = series_no.value_counts().head(30).sort_values()
```

```python
m_yes.plot(kind='barh',figsize=(5,10), title='Palavras de quem votou pelo impeachment')
```

```python
m_no.plot(kind='barh',figsize=(5,10), title='Palavras de quem votou contra o impeachment')
```

```python
sample_votes
```

```python
m_yes
```

```python
m_no
```

```python
rs = RSLPStemmer()
for i in range(len(token_yes)):
    token_yes[i] = rs.stem(token_yes[i])

for i in range(len(token_no)):
    token_no[i] = rs.stem(token_no[i])
```

```python
#fazer uma funcao para tratamento dos textos texto por texto para depois fazer a analise nas 30 mais frequentes
def labels(column):
    speech = column['Fala']
    vote = column['Voto']
    token = word_tokenize(speech)
    words = []
    if vote == 'Sim':
        for word in token:
            if word in m_yes.index:
                words.append(word)
    if vote == 'Não':
        for word in token:
            if word in m_no.index:
                words.append(word)
    return words
```

```python
sample_votes.head()
```

```python
label = sample_votes.apply(labels, axis=1)
```

```python
label
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
data = mlb.fit_transform(label)
columns = mlb.classes_
label_2 = pd.DataFrame(data=data, columns=columns, index=sample_votes.index)
data_2 = pd.concat([sample_votes, label_2], axis=1)
```

```python
data
```

```python
columns
```

```python
label_2
```

```python
data_2
```

```python
X = data_2.drop(['Partido','Estado','Voto','Gênero','Fala'], axis=1)
y = data_2.Voto.astype('category').cat.codes #transformar em dados categoricos (1 sendo spam, e 0 para ham)
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(verbose=1)
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
X_train, X_test, y_train, y_test = train_test_split(X,y)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
